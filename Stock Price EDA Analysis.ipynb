```python
df_telsa=pdr.get_data_yahoo('TSLA')
df_telsa
```

```python
df_google=pdr.get_data_yahoo('GOOGL')
df_google
```

```python
df_apple=pdr.get_data_yahoo('AAPL')
df_apple
```

```python
df_telsa['Open'].plot(figsize=(15,5))
```

```python
df_amazon['Open'].plot(figsize=(15,5))
```

```python
df_google=df_google.tail(365)
df_google
```

```python
df_amzon=df_amazon.tail(365)
df_amazon
```

```python
df_apple=df_amazon.tail(365)
df_apple
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(df_google.index,df_google.High,color='r',label='Google')
plt.plot(df_amazon.index,df_amazon.High,color='g',label='Amazon')
plt.plot(df_apple.index,df_apple.High,color='b',label='Apple')
```

```python
df_telsa['30 Rolling']=df_telsa['High'].rolling(30).mean()
```

```python
df_telsa[['High','30 Rolling']]
```

```python
df_telsa['20 Rolling']=df_telsa['High'].rolling(20).mean()
```

```python
df_telsa['10 Rolling']=df_telsa['High'].rolling(10).mean()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_telsa.index,df_telsa['High'],color='r',label='Original')
plt.plot(df_telsa.index,df_telsa['20 Rolling'],color='g',label='20 Rolling')
plt.plot(df_telsa.index,df_telsa['10 Rolling'],color='b',label='10 Rolling')
plt.plot(df_telsa.index,df_telsa['30 Rolling'],color='b',label='30 Rolling')
```
