```python
import pandas as pd
df=pd.read_csv('dataset/Pokemon.csv')
```

```python
df
```

```python
#lamdba with constant values
#apply + lamdba

df['newcol']=df.Speed.apply(lambda x:x*10)
```

```python
df.head()
```

```python
def myfun(x):
    return x+10

df['newcol1']=df.HP.apply(myfun)
```

```python
df.head()
```

```python
#lambda with condition
#>50 0
#<50 1

df['newcol2']=df['Defense'].apply(lambda x: 0 if x>50 else 1 )
```

```python
df.head(2)
```

```python
#total > 300 0
# <300 1
df['newcol3'] = df["Total"].apply(lambda x:1 if x<300 else 1)
```

```python
df.head()
```

```python
df.drop('newcol',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.drop(['newcol1','newcol2','newcol3'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
#total 
#<100 -1
#> 100 0
# >200 1
# >300 2
#userdefined function
#lambda cannot be used if the condtion is more than 2

def changeValue(value):
    if value<100:
        return -1
    elif value>300:
        return 0
    elif value>200:
        return 1
    elif value>100:
        return 2
    

df['newcol3'] = df["Total"].apply(changeValue)
```

```python
df.head()
```

```python
df['newcol3'].value_counts()
```

```python
df['meanvalue']=df[['HP','Total','Attack']].apply(lambda x: (x[0]+x[1]+x[2])/3,axis=1)

```

```python
df.head()
```

```python
#Speed + HP >100 = 0
#           <100 =1

df['newcol']=df[['Speed','HP']].apply(lambda x: 0 if x[0]+x[1]>100 else 1,axis=1)
df.head()
```

```python
def changeFun(x):
    if x[0]+x[1]>100:
        return 0
    else:
        return 1
    
df['newcol']=df[['Speed','HP']].apply(changeFun,axis=1)
df.head()
```
