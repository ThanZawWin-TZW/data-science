```python
#File Combine

import pandas as pd
movies=pd.read_csv('dataset/movies/movies.csv')
movies.head()
```

```python
ratings=pd.read_csv('dataset/movies/ratings.csv')
ratings.head()
```

```python
tags=pd.read_csv('dataset/movies/tags.csv')
tags.head()
```

```python
#merge, join

#moives , tags, ratings
all_files=movies.merge(tags,on='movieId',how='inner').merge(ratings,on='movieId',how='inner')
all_files.head()
```

```python
all_files['Comedy']=all_files['genres'].apply(lambda x: 'Comedy' if 'Comedy' in x else "No Comedy")
```

```python
all_files.groupby(['Comedy'])['rating'].mean()
```

```python
"""all_files['new_col']

Comedy - comedy
drama drama
action action"""
def changeValue(x):
    if 'Drama' in x:
        return 'Drama'
    elif 'Comedy' in x:
        return 'Comedy'
    elif 'Action' in x:
        return 'Action'
    else:
        return 'Other'

all_files['newcol']=all_files['genres'].apply(changeValue)
```

```python
all_files.newcol.value_counts().plot(kind='bar')
```

```python
all_files.groupby('newcol')['rating'].mean().plot(kind='bar',color=['r','g','b','y'])
```

```python
import os
files=os.listdir('dataset/Sales_Data')
```

```python
all_df=pd.DataFrame() #empty
for x in files:
    f=pd.read_csv('dataset/Sales_Data/'+x)
    all_df=pd.concat([all_df,f])
    
```

```python
all_df
```

```python
all_df.isnull().sum()
```

```python
all_df.dropna(inplace=True)
```

```python
all_df.dtypes
```

```python
all_df=all_df[all_df['Order Date']!='Order Date']
```

```python
all_df['Quantity Ordered']=pd.to_numeric(all_df['Quantity Ordered'])
```

```python
all_df['Price Each']=pd.to_numeric(all_df['Price Each'])
```

```python
all_df['Order Date']=pd.to_datetime(all_df['Order Date'])
```

```python
all_df['Month']=all_df['Order Date'].dt.month_name()
```

```python
all_df['Day']=all_df['Order Date'].dt.day_of_week
```

```python
all_df['City']=all_df['Purchase Address'].apply(lambda x:x.split(',')[1].strip())
```

```python
all_df
```

```python
#Quantity Ordered = int
#Price Each - int
#Order Date - date, newcol- Month, Day
#Purchase Address - City

all_df.to_csv("dataset/Sales_Data/readyfile.csv")
```
